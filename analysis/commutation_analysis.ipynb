```python
import sys
sys.path.append("..")


from pytket import OpType, Circuit
from canopus.utils import tket_to_qiskit
from canopus.synthesis import rebase_to_tk2
import regulus
from pytket.circuit.display import render_circuit_jupyter
from pytket.utils import compare_unitaries
import pytket.qasm
import pytket.passes
import cirq
from math import pi

from regulus.utils.passes import approximately_commutative
from regulus import gates
```

```python
import numpy as np

for _ in range(100):
    thetas = np.random.rand(3) * 2 * pi
    phis = np.random.rand(3) * 2 * pi
    g1 = gates.Can(*thetas).on([0,1])
    g2 = gates.Can(*phis).on([1,2])

    if approximately_commutative(g1, g2):
        print("Commutative")
```

```python
thetas = [pi/2, 0, 0]
phis = [pi/2, 0, 0]
```

```python
c1 = Circuit(3)
c1.TK2(*thetas, 0, 1)
c1.TK2(*phis, 1, 2)
tket_to_qiskit(c1).draw()
```

```python
c2 = Circuit(3)
c2.TK2(*phis, 1, 2)
c2.TK2(*thetas, 0, 1)
tket_to_qiskit(c2).draw()
```

```python
compare_unitaries(c1.get_unitary(), c2.get_unitary())
```
