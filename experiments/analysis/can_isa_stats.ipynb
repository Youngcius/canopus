```python
import sys
sys.path.append('../..')

from qiskit import QuantumCircuit
from collections import Counter
import canopus
import os
import numpy as np
from collections import Counter

from regulus.utils.functions import replace_close_to_zero_with_zero
```

统计 Can-XX 以及 Can-XY 情况下的 Canonical coordinates

```python
can_xx_dpath = '../output/canopus/chain/can_xx'
can_xy_dpath = '../output/canopus/chain/can_xy'

fnames = os.listdir(can_xx_dpath)
```

```python
stats = Counter()
for fname in fnames:
    if 'qft' in fname or 'ising' in fname or 'dnn' in fname or 'qpe' in fname or 'gcm' in fname:
        continue
    print(fname)
    qc = QuantumCircuit.from_qasm_file(os.path.join(can_xx_dpath, fname))
    qc = canopus.rebase_to_canonical(qc)
    print(Counter([canopus.utils.replace_close_to_zero_with_zero(instr.operation.params) for instr in qc.data if isinstance(instr.operation, canopus.CanonicalGate)]))
```
