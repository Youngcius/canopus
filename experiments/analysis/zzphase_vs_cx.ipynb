```python
import os
import sys
sys.path.append('../..')

import canopus
import numpy as np
from natsort import natsorted
from qiskit import QuantumCircuit
```

**TODO: 为什么ZZPhase结果比CX要差**

```python
cx_output_dpath = '../output/sabre/chain/cx'
zzphase_output_dpath = '../output/sabre/chain/zzphase'

cx_synth_estimator = canopus.SynthCostEstimator(isa_type='cx')
zzphase_synth_estimator = canopus.SynthCostEstimator(isa_type='zzphase')
```

```python
fnames = natsorted(os.listdir(cx_output_dpath))

cx_costs = []
zzphase_costs = []
for fname in fnames:
    cx_fname = os.path.join(cx_output_dpath, fname)
    zzphase_fname = os.path.join(zzphase_output_dpath, fname)

    qc_cx = QuantumCircuit.from_qasm_file(cx_fname)
    qc_zzphase = QuantumCircuit.from_qasm_file(zzphase_fname)
    cx_costs.append(cx_synth_estimator.eval_circuit_duration(qc_cx))
    zzphase_costs.append(zzphase_synth_estimator.eval_circuit_duration(qc_zzphase))
```

```python
(np.array(cx_costs) / np.array(zzphase_costs)).round(3)
```
